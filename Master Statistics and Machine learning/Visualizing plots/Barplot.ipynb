# Importing libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Data 

```python
rows = 30
col = 6

data = np.zeros((rows, col))

for i in range(col):
    data[:,i] = 30*np.random.randn(rows) * (2*i/(col-1)-1)**2 + (i+1)**2
```

# Plotting 

```python
fig, ax = plt.subplots(1,3, figsize=(7,3))

ax[0].bar(range(col), np.mean(data, axis=0))
ax[0].set_title('Bar plot')

ax[1].errorbar(range(col), np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='s', linestyle='')
ax[1].set_title('Errorbar plot')

ax[2].bar(range(col), np.mean(data, axis=0))
ax[2].errorbar(range(col), np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='s', linestyle='')
ax[2].set_title('Error +  bar plot')
```

# Manually specify x-axis coordinates 

```python
xcrossings = [1,2,4,5,6,9]

plt.bar(xcrossings, np.mean(data, axis=0))
plt.errorbar(xcrossings, np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='.', linestyle='')
plt.title('Bar + Error plot')

plt.show()
```

# Using pandas 

```python
m = [[2,5,4,3], [1,1,8,6]]

fig, ax = plt.subplots(2,2, figsize=(8,8))

ax[0,0].imshow(m)

df = pd.DataFrame(m,columns=['prop 0', 'prop 1', 'prop 2' 'prop 3', 'prop 4'])
df.plot(ax=ax[1,0], kind='bar')
ax[1,0].set_title('Gropu by rows')

ax[0,1].imshow(np.array(m).T)
df.T.plot(ax=ax[1,1], kind='bar')
ax[1,1].set_title('Grouping by columns')

plt.show()
```
