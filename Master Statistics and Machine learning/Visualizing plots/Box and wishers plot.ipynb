# Importing libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data 

```python
rows = 30
col = 6

data = np.zeros((rows, col))

for i in range(col):
    data[:,i] = 30*np.random.randn(rows) * (2*i/(col-1)-1)**2 + (i+1)**2
```

```python
data.shape
```

# Plot using matplotlib 

```python
plt.boxplot(data)
plt.show()
```

# Plot using seaborn 

```python
sns.boxplot(data, orient='v')
plt.show()
```

```python
df = pd.DataFrame(data, columns=['One', 'Two', 'Three', 'Four', 'Five', 'Six'])
sns.boxplot(data, orient='h')
plt.show()
```
