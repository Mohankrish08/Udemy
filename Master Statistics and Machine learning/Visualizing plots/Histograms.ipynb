# Importing libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Data  

```python
n = 1000

data = np.exp(np.random.randn(n)/2)
```

```python
plt.plot(data, 's')
```

# Visualizing with histogram 

```python
k = 40

plt.hist(data, bins=k)
plt.show()
```

```python
y,x = np.histogram(data, bins=k)

xx = (x[1:]+x[:-1]) / 2

plt.plot(xx, y)
plt.show()
```
